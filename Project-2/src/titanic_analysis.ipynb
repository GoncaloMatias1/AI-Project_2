### Step 4: Finding Outsiders

### Outlier Detection in the Titanic Dataset

We will now check for outliers in all numerical columns using the IQR method. The following code finds outliers for each specified numerical column and generates scatter plots and histograms to visualize these outliers.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to find outliers using the IQR method
def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Identify numerical columns in the current dataset
numerical_columns = df.select_dtypes(include=['number']).columns.tolist()

# Dictionary to hold outliers for each column
outliers_dict = {}

# Find outliers for each numerical column
for column in numerical_columns:
    outliers_dict[column] = find_outliers_iqr(df, column)
    print(f"Number of {column} outliers: {len(outliers_dict[column])}")

    # Scatter plot for outliers
    plt.figure(figsize=(10, 6))
    plt.scatter(df.index, df[column], color='blue', label='Data Points')
    plt.scatter(outliers_dict[column].index, outliers_dict[column][column], color='red', label='Outliers')
    plt.axhline(y=df[column].quantile(0.25) - 1.5 * (df[column].quantile(0.75) - df[column].quantile(0.25)), color='green', linestyle='--', label='Lower Bound')
    plt.axhline(y=df[column].quantile(0.75) + 1.5 * (df[column].quantile(0.75) - df[column].quantile(0.25)), color='purple', linestyle='--', label='Upper Bound')
    plt.title(f'Scatter Plot of {column.capitalize()} with Outliers')
    plt.xlabel('Index')
    plt.ylabel(column.capitalize())
    plt.legend()
    plt.show()

    # Histogram for outliers
    plt.figure(figsize=(10, 6))
    plt.hist(df[column].dropna(), bins=30, color='blue', edgecolor='black', alpha=0.7)
    plt.axvline(x=df[column].quantile(0.25) - 1.5 * (df[column].quantile(0.75) - df[column].quantile(0.25)), color='green', linestyle='--', label='Lower Bound')
    plt.axvline(x=df[column].quantile(0.75) + 1.5 * (df[column].quantile(0.75) - df[column].quantile(0.25)), color='purple', linestyle='--', label='Upper Bound')
    plt.title(f'Histogram of {column.capitalize()} with Outlier Thresholds')
    plt.xlabel(column.capitalize())
    plt.ylabel('Count')
    plt.legend()
    plt.show()
```

### Step 3: Correlation Analysis

To identify relationships between different variables in the dataset, we will compute and visualize the correlation matrix. This will help us understand which features are most important for predicting the target variable (`survived`) and inform our data processing decisions.

We will start by encoding categorical variables and dropping columns that are not suitable for correlation analysis.    


```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Print column names to check for any discrepancies
print("Columns in the DataFrame before encoding and dropping:")
print(df.columns)

# Encode categorical variables
label_encoder = LabelEncoder()
df['sex'] = label_encoder.fit_transform(df['sex'])
df['embarked'] = label_encoder.fit_transform(df['embarked'].astype(str))

# Define columns to drop
columns_to_drop = ['name', 'ticket', 'cabin', 'boat','home.dest']

# Ensure the columns to drop are in the DataFrame
columns_to_drop = [col for col in columns_to_drop if col in df.columns]

# Drop non-numeric columns
df = df.drop(columns=columns_to_drop)

# Verify that only numeric columns remain
print("Columns after dropping non-numeric ones:")
print(df.columns)

# Compute the correlation matrix
correlation_matrix = df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)

# Set titles and labels
plt.title('Correlation Matrix of Titanic Dataset')
plt.show()
```
