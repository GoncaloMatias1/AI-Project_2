### Step 4: Finding Outsiders

### Outlier Detection in the Titanic Dataset

Outliers can significantly impact statistical analyses and can be indicative of data entry errors, or they could be accurate representations of rare cases. We will explore outliers in the 'age' and 'fare' columns using the IQR method, Z-scores, and visualization techniques.


```python
# Function to find outliers using the IQR method
def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Find outliers in 'age' and 'fare' columns
age_outliers = find_outliers_iqr(df, 'age')
fare_outliers = find_outliers_iqr(df, 'fare')

print(f"Number of age outliers: {len(age_outliers)}")
print(f"Number of fare outliers: {len(fare_outliers)}")

# Scatter plot for age outliers
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['age'], color='blue', label='Data Points')
plt.scatter(age_outliers.index, age_outliers['age'], color='red', label='Outliers')
plt.axhline(y=df['age'].quantile(0.25) - 1.5 * (df['age'].quantile(0.75) - df['age'].quantile(0.25)), color='green', linestyle='--', label='Lower Bound')
plt.axhline(y=df['age'].quantile(0.75) + 1.5 * (df['age'].quantile(0.75) - df['age'].quantile(0.25)), color='purple', linestyle='--', label='Upper Bound')
plt.title('Scatter Plot of Age with Outliers')
plt.xlabel('Index')
plt.ylabel('Age')
plt.legend()
plt.show()

# Scatter plot for fare outliers
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['fare'], color='blue', label='Data Points')
plt.scatter(fare_outliers.index, fare_outliers['fare'], color='red', label='Outliers')
plt.axhline(y=df['fare'].quantile(0.25) - 1.5 * (df['fare'].quantile(0.75) - df['fare'].quantile(0.25)), color='green', linestyle='--', label='Lower Bound')
plt.axhline(y=df['fare'].quantile(0.75) + 1.5 * (df['fare'].quantile(0.75) - df['fare'].quantile(0.25)), color='purple', linestyle='--', label='Upper Bound')
plt.title('Scatter Plot of Fare with Outliers')
plt.xlabel('Index')
plt.ylabel('Fare')
plt.legend()
plt.show()

# Histogram for age with outlier thresholds
plt.figure(figsize=(10, 6))
plt.hist(df['age'].dropna(), bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.axvline(x=df['age'].quantile(0.25) - 1.5 * (df['age'].quantile(0.75) - df['age'].quantile(0.25)), color='green', linestyle='--', label='Lower Bound')
plt.axvline(x=df['age'].quantile(0.75) + 1.5 * (df['age'].quantile(0.75) - df['age'].quantile(0.25)), color='purple', linestyle='--', label='Upper Bound')
plt.title('Histogram of Age with Outlier Thresholds')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()
plt.show()

# Histogram for fare with outlier thresholds
plt.figure(figsize=(10, 6))
plt.hist(df['fare'], bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.axvline(x=df['fare'].quantile(0.25) - 1.5 * (df['fare'].quantile(0.75) - df['fare'].quantile(0.25)), color='green', linestyle='--', label='Lower Bound')
plt.axvline(x=df['fare'].quantile(0.75) + 1.5 * (df['fare'].quantile(0.75) - df['fare'].quantile(0.25)), color='purple', linestyle='--', label='Upper Bound')
plt.title('Histogram of Fare with Outlier Thresholds')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.legend()
plt.show()
```

### Step 3: Correlation Analysis

To identify relationships between different variables in the dataset, we will compute and visualize the correlation matrix. This will help us understand which features are most important for predicting the target variable (`survived`) and inform our data processing decisions.

We will start by encoding categorical variables and dropping columns that are not suitable for correlation analysis.    


```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Print column names to check for any discrepancies
print("Columns in the DataFrame before encoding and dropping:")
print(df.columns)

# Encode categorical variables
label_encoder = LabelEncoder()
df['sex'] = label_encoder.fit_transform(df['sex'])
df['embarked'] = label_encoder.fit_transform(df['embarked'].astype(str))

# Define columns to drop
columns_to_drop = ['name', 'ticket', 'cabin', 'boat', 'body', 'home.dest']

# Ensure the columns to drop are in the DataFrame
columns_to_drop = [col for col in columns_to_drop if col in df.columns]

# Drop non-numeric columns
df = df.drop(columns=columns_to_drop)

# Verify that only numeric columns remain
print("Columns after dropping non-numeric ones:")
print(df.columns)

# Compute the correlation matrix
correlation_matrix = df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)

# Set titles and labels
plt.title('Correlation Matrix of Titanic Dataset')
plt.show()
```
