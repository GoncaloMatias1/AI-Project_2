### 3 - Finding Outsiders

### Outlier Detection in the Titanic Dataset

We will now check for outliers in all numerical columns using the IQR method. The following code finds outliers for each specified numerical column and generates scatter plots and histograms to visualize these outliers.


````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

file_path = '../data/Titanic Dataset.csv'
df = pd.read_csv(file_path)

df['age'] = df['age'].fillna(df['age'].median())
df['fare'] = df['fare'].fillna(df['fare'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

label_encoders = {}
categorical_columns = ['sex', 'embarked']

for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column].astype(str))

def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

numerical_columns = ['pclass', 'survived', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']

outliers_dict = {}

for column in numerical_columns:
    outliers_dict[column] = find_outliers_iqr(df, column)
    print(f"Number of {column} outliers: {len(outliers_dict[column])}")

    plt.figure(figsize=(10, 6))
    plt.scatter(df.index, df[column], color='blue', label='Data Points')
    plt.scatter(outliers_dict[column].index, outliers_dict[column][column], color='red', label='Outliers')
    plt.axhline(y=df[column].quantile(0.25) - 1.5 * (df[column].quantile(0.75) - df[column].quantile(0.25)), color='green', linestyle='--', label='Lower Bound')
    plt.axhline(y=df[column].quantile(0.75) + 1.5 * (df[column].quantile(0.75) - df[column].quantile(0.25)), color='purple', linestyle='--', label='Upper Bound')
    plt.title(f'Scatter Plot of {column.capitalize()} with Outliers')
    plt.xlabel('Index')
    plt.ylabel(column.capitalize())
    plt.legend()
    plt.show()

    plt.figure(figsize=(10, 6))
    plt.hist(df[column].dropna(), bins=30, color='blue', edgecolor='black', alpha=0.7)
    plt.axvline(x=df[column].quantile(0.25) - 1.5 * (df[column].quantile(0.75) - df[column].quantile(0.25)), color='green', linestyle='--', label='Lower Bound')
    plt.axvline(x=df[column].quantile(0.75) + 1.5 * (df[column].quantile(0.75) - df[column].quantile(0.25)), color='purple', linestyle='--', label='Upper Bound')
    plt.title(f'Histogram of {column.capitalize()} with Outlier Thresholds')
    plt.xlabel(column.capitalize())
    plt.ylabel('Count')
    plt.legend()
    plt.show()
````

`Pclass`, `Survived`, `Sex`, `Embarked`: No outliers were detected in these columns, indicating a well-defined categorical distribution.

`Age`, `Sibsp`, `Parch`, `Fare`: These columns have a significant number of outliers. For instance, the Age column has 101 outliers, Sibsp has 57, Parch has 307, and Fare has 171.

For `Age`, `Sibsp`, `Parch`, and `Fare`, the presence of outliers might affect the model's performance.

To ensure that the extreme values in our dataset do not adversely affect our analysis and model performance, we will handle the outliers in the `Age`, `Sibsp`, `Parch`, and `Fare` columns by capping them. Capping involves setting a limit on the values that exceed a certain threshold, which helps in reducing the skewness and making the data more robust.


````python
import pandas as pd

file_path = '../data/Titanic Dataset.csv'
df = pd.read_csv(file_path)

def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series.clip(lower_bound, upper_bound)

df['age'] = cap_outliers(df['age'])
df['sibsp'] = cap_outliers(df['sibsp'])
df['parch'] = cap_outliers(df['parch'])
df['fare'] = cap_outliers(df['fare'])

def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

print("Number of age outliers after capping:", len(find_outliers_iqr(df, 'age')))
print("Number of sibsp outliers after capping:", len(find_outliers_iqr(df, 'sibsp')))
print("Number of parch outliers after capping:", len(find_outliers_iqr(df, 'parch')))
print("Number of fare outliers after capping:", len(find_outliers_iqr(df, 'fare')))
````

The results above indicate that after applying the capping method using the Interquartile Range (IQR), all outliers in the columns 'age', 'sibsp', 'parch', and 'fare' have been successfully capped. This ensures that the data within these columns now fall within a specified range, eliminating the influence of extreme values and providing a more robust dataset for further analysis and modeling.

## Step 5 : Feature Engineering

Based on the correlation matrix, the most important step to proceed with is Feature Engineering. This involves creating new features that might provide more predictive power for the model.

Combining Features and Binning are two powerful techniques in feature engineering. Combining features such as `sibsp` and `parch` to create a family size feature can be particularly useful. Additionally, binning the `age` variable into age groups can help in capturing non-linear relationships more effectively.

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

file_path = '../data/Titanic Dataset.csv'  
df = pd.read_csv(file_path)

columns_to_drop = ['name', 'ticket', 'home.dest', 'body', 'cabin', 'boat']
df = df.drop(columns=columns_to_drop)

df['age'] = df['age'].fillna(df['age'].median())
df['fare'] = df['fare'].fillna(df['fare'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

label_encoders = {}
categorical_columns = ['sex', 'embarked']

for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column].astype(str))

def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series.clip(lower_bound, upper_bound)

df['age'] = cap_outliers(df['age'])
df['sibsp'] = cap_outliers(df['sibsp'])
df['parch'] = cap_outliers(df['parch'])
df['fare'] = cap_outliers(df['fare'])

df['family_size'] = df['sibsp'] + df['parch'] + 1  

age_bins = [0, 12, 18, 35, 60, np.inf]
age_labels = ['Child', 'Teenager', 'Adult', 'Senior', 'Elder']
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels)

label_encoders['age_group'] = LabelEncoder()
df['age_group'] = label_encoders['age_group'].fit_transform(df['age_group'].astype(str))

df.drop(columns=['sibsp', 'parch'], inplace=True)

print(df.head())

numerical_columns = ['pclass', 'survived', 'sex', 'age', 'fare', 'embarked', 'family_size', 'age_group']

correlation_matrix = df[numerical_columns].corr()
plt.figure(figsize=(12, 10))
plt.title('Correlation Matrix after Feature Engineering')
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
````

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score, confusion_matrix
import time
import seaborn as sb
import matplotlib.pyplot as plt

dt_md = DecisionTreeClassifier(max_depth=5, random_state=42)

start_time = time.time()
dt_md.fit(X_train, y_train)
train_time_dt = time.time() - start_time

start_time = time.time()
dt_train_pred = dt_md.predict(X_train)
train_pred_time_dt = time.time() - start_time

start_time = time.time()
dt_test_pred = dt_md.predict(X_test)
test_pred_time_dt = time.time() - start_time

dt_precision = precision_score(y_test, dt_test_pred) * 100
dt_recall = recall_score(y_test, dt_test_pred) * 100
dt_accuracy = accuracy_score(y_test, dt_test_pred) * 100
dt_f1 = f1_score(y_test, dt_test_pred) * 100

print(f'Precision Score: {dt_precision:.2f} %')
print(f'Recall Score: {dt_recall:.2f} %')
print(f'Accuracy Score: {dt_accuracy:.2f} %')
print(f'F1 Score: {dt_f1:.2f} %')
print(f'Training Time: {train_time_dt:.2f} seconds')
print(f'Testing Time: {test_pred_time_dt:.2f} seconds')

dt_cm_train = confusion_matrix(y_train, dt_train_pred)
sb.heatmap(dt_cm_train, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Training')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()

dt_cm_test = confusion_matrix(y_test, dt_test_pred)
sb.heatmap(dt_cm_test, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Test')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()

from sklearn import tree
plt.figure(figsize=(20, 10))
tree.plot_tree(dt_md, filled=True, fontsize=10, rounded=True)
plt.show()
````

#### 2 - K-Nearest Neighbors

````python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import time

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn = KNeighborsClassifier(n_neighbors=5)

start_time = time.time()
knn.fit(X_train_scaled, y_train)
train_time_knn = time.time() - start_time

start_time = time.time()
y_pred_knn = knn.predict(X_test_scaled)
test_time_knn = time.time() - start_time

accuracy_knn = accuracy_score(y_test, y_pred_knn) * 100
precision_knn = precision_score(y_test, y_pred_knn) * 100
recall_knn = recall_score(y_test, y_pred_knn) * 100
f1_knn = f1_score(y_test, y_pred_knn) * 100
report_knn = classification_report(y_test, y_pred_knn)
conf_matrix_knn = confusion_matrix(y_test, y_pred_knn)

print("K-Nearest Neighbors Classifier")
print(f"Accuracy: {accuracy_knn:.2f}%")
print(f"Precision: {precision_knn:.2f}%")
print(f"Recall: {recall_knn:.2f}%")
print(f"F1 Score: {f1_knn:.2f}%")
print("Classification Report:")
print(report_knn)
print(f'Training Time: {train_time_knn:.2f} seconds')
print(f'Testing Time: {test_time_knn:.2f} seconds')

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_knn, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Survived', 'Survived'], yticklabels=['Not Survived', 'Survived'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - KNN')
plt.show()
````

#### 3 - Support Vector Machine

````python
import numpy as np
import time
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns

svm_model = SVC(kernel='linear', random_state=42)

start_time = time.time()
svm_model.fit(X_train, y_train)
train_time_svm = time.time() - start_time

start_time = time.time()
y_pred_svm = svm_model.predict(X_test)
test_time_svm = time.time() - start_time

accuracy_svm = accuracy_score(y_test, y_pred_svm) * 100
precision_svm = precision_score(y_test, y_pred_svm) * 100
recall_svm = recall_score(y_test, y_pred_svm) * 100
f1_svm = f1_score(y_test, y_pred_svm) * 100
report_svm = classification_report(y_test, y_pred_svm)
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)

print("Support Vector Machine (SVM)")
print(f"Accuracy: {accuracy_svm:.2f}%")
print(f"Precision: {precision_svm:.2f}%")
print(f"Recall: {recall_svm:.2f}%")
print(f"F1 Score: {f1_svm:.2f}%")
print("Classification Report:")
print(report_svm)
print(f'Training Time: {train_time_svm:.2f} seconds')
print(f'Testing Time: {test_time_svm:.2f} seconds')

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_svm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Survived', 'Survived'], yticklabels=['Not Survived', 'Survived'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - SVM')
plt.show()
````

#### 4 - Neural Network

````python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
import time
import matplotlib.pyplot as plt
import seaborn as sns

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)

model = Sequential()
model.add(Dense(64, input_shape=(X_train_scaled.shape[1],), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

start_time = time.time()
history = model.fit(X_train_scaled, y_train_categorical, epochs=50, batch_size=32, validation_split=0.2, verbose=1)
train_time_nn = time.time() - start_time

start_time = time.time()
y_pred_proba = model.predict(X_test_scaled)
y_pred = np.argmax(y_pred_proba, axis=1)
test_time_nn = time.time() - start_time

accuracy_nn = accuracy_score(y_test, y_pred) * 100
precision_nn = precision_score(y_test, y_pred) * 100
recall_nn = recall_score(y_test, y_pred) * 100
f1_nn = f1_score(y_test, y_pred) * 100
report_nn = classification_report(y_test, y_pred)
conf_matrix_nn = confusion_matrix(y_test, y_pred)

print("Neural Network Classifier")
print(f"Accuracy: {accuracy_nn:.2f}%")
print(f"Precision: {precision_nn:.2f}%")
print(f"Recall: {recall_nn:.2f}%")
print(f"F1 Score: {f1_nn:.2f}%")
print("Classification Report:")
print(report_nn)
print(f'Training Time: {train_time_nn:.2f} seconds')
print(f'Testing Time: {test_time_nn:.2f} seconds')

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_nn, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Survived', 'Survived'], yticklabels=['Not Survived', 'Survived'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Neural Network')
plt.show()
````

The F1 Score is the harmonic mean of precision and recall, providing a single metric that balances both concerns. Support Vector Machine and K-Nearest Neighbors have higher F1 Scores compared to Decision Tree and Neural Network, indicating a balanced performance between precision and recall.

Training time measures how long each algorithm takes to learn from the training data. All algorithms have relatively low training times, with Decision Tree being slightly faster.
